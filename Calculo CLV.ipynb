```python
import pandas as pd

def explorar_dataset(nombre, data):
    print(f"\n\n############################")
    print(f"### EXPLORANDO DATASET: {nombre}")
    print(f"############################")
    
    # 0. Eliminar columnas duplicadas (CRÍTICO)
    if data.columns.duplicated().any():
        print("\n⚠️ Se detectaron columnas duplicadas. Se eliminarán automáticamente.")
        print("Columnas duplicadas:", data.columns[data.columns.duplicated()].tolist())
        data = data.loc[:, ~data.columns.duplicated()]
    
    # 1. Ver columnas disponibles
    print("\n====== COLUMNAS DEL DATASET ======")
    print(data.columns)
    
    # 2. SEGMENTO_2026
    if "SEGMENTO_2026" in data.columns:
        print("\n====== SEGMENTOS DISPONIBLES EN SEGMENTO_2026 ======")
        
        col = data["SEGMENTO_2026"]
        
        # Manejo por si aún viene como DataFrame
        if isinstance(col, pd.DataFrame):
            print("⚠️ SEGMENTO_2026 viene como DataFrame, tomando primera columna")
            col = col.iloc[:, 0]
        
        print(col.dropna().unique())
        
        # 3. Conteo de clientes únicos por segmento
        if "CLIENTE_ID" in data.columns:
            cliente_col = data["CLIENTE_ID"]
            
            if isinstance(cliente_col, pd.DataFrame):
                print("⚠️ CLIENTE_ID viene como DataFrame, tomando primera columna")
                cliente_col = cliente_col.iloc[:, 0]
            
            clientes_unicos = (
                pd.DataFrame({
                    "SEGMENTO_2026": col,
                    "CLIENTE_ID": cliente_col
                })
                .dropna(subset=["SEGMENTO_2026", "CLIENTE_ID"])
                .groupby("SEGMENTO_2026")["CLIENTE_ID"]
                .nunique()
                .reset_index(name="Clientes_Unicos")
            )
            
            print("\n====== CANTIDAD DE CLIENTES ÚNICOS POR SEGMENTO ======")
            print(clientes_unicos)
        else:
            print("\n⚠️ La columna 'CLIENTE_ID' no está en el archivo.")
    else:
        print("\n⚠️ La columna 'SEGMENTO_2026' no está en el archivo.")
    
    # 4. CONVERGENCIA_2
    if "CONVERGENCIA_2" in data.columns:
        print("\n====== VALORES ÚNICOS EN CONVERGENCIA_2 ======")
        
        col = data["CONVERGENCIA_2"]
        if isinstance(col, pd.DataFrame):
            col = col.iloc[:, 0]
        
        print(col.dropna().unique())
    else:
        print("\n⚠️ La columna 'CONVERGENCIA_2' no está en el archivo.")
    
    # 5. LINEA
    if "LINEA" in data.columns:
        print("\n====== VALORES ÚNICOS EN LINEA ======")
        
        col = data["LINEA"]
        if isinstance(col, pd.DataFrame):
            col = col.iloc[:, 0]
        
        print(col.dropna().unique())
    else:
        print("\n⚠️ La columna 'LINEA' no está en el archivo.")


# ============================
# EJECUCIÓN
# ============================

explorar_dataset("df", df)
explorar_dataset("d2", d2)
```

```python
# ==========================
# LIMPIEZA CONVERGENCIA + AJUSTE NUMÉRICO (VERSIÓN FINAL)
# ==========================

import pandas as pd
import numpy as np

col_conv2 = "CONVERGENCIA_2"
col_conv = "CONVERGENCIA"

if col_conv2 in df.columns and col_conv in df.columns:
    
    print("\n🧹 Iniciando limpieza integral de convergencia...")
    
    # 1. Asegurar tipos correctos
    df[col_conv2] = df[col_conv2].astype("string")
    df[col_conv] = pd.to_numeric(df[col_conv], errors="coerce")
    
    # 2. Máscara: valores exactamente "SIN_NIVEL"
    mask_sin_nivel_puro = df[col_conv2] == "SIN_NIVEL"
    
    # 3. Máscara: combinaciones que terminan en "-SIN_NIVEL"
    mask_sin_nivel_sufijo = df[col_conv2].str.endswith("-SIN_NIVEL", na=False)
    
    # ==========================
    # CASO 1: "SIN_NIVEL" → NULL en ambas columnas
    # ==========================
    df.loc[mask_sin_nivel_puro, col_conv2] = pd.NA
    df.loc[mask_sin_nivel_puro, col_conv] = pd.NA
    
    # ==========================
    # CASO 2: "ALGO-SIN_NIVEL" → quitar sufijo y ajustar convergencia
    # ==========================
    # Quitar el sufijo -SIN_NIVEL
    df.loc[mask_sin_nivel_sufijo, col_conv2] = (
        df.loc[mask_sin_nivel_sufijo, col_conv2]
        .str.replace(r"-SIN_NIVEL$", "", regex=True)
        .str.strip()
    )
    
    # Restar 1 a la convergencia donde aplique
    df.loc[mask_sin_nivel_sufijo, col_conv] = (
        df.loc[mask_sin_nivel_sufijo, col_conv] - 1
    )
    
    # ==========================
    # CASO 3: Si convergencia queda en 0 → NULL (como pediste)
    # ==========================
    mask_conv_cero = df[col_conv] == 0
    df.loc[mask_conv_cero, col_conv] = pd.NA
    
    # ==========================
    # Limpieza final
    # ==========================
    df[col_conv2] = df[col_conv2].str.strip()
    
    # ==========================
    # Validación final
    # ==========================
    print("\n====== VALORES ÚNICOS LIMPIOS EN CONVERGENCIA_2 ======")
    print(sorted(df[col_conv2].dropna().unique()))
    
    print("\n📊 Resumen después de limpieza:")
    print("Nulos en CONVERGENCIA_2:", df[col_conv2].isna().sum())
    print("Nulos en CONVERGENCIA:", df[col_conv].isna().sum())
    print("Registros con sufijo SIN_NIVEL ajustados:", mask_sin_nivel_sufijo.sum())

else:
    print("⚠️ Columnas CONVERGENCIA_2 o CONVERGENCIA no encontradas")
```

```python
# ==========================
# AJUSTE: LINEA NULL → RESTAR 1 A NUM_LINEAS
# ==========================

col_linea = "LINEA"
col_num_lineas = "NUM_LINEAS"

if col_linea in df.columns and col_num_lineas in df.columns:
    
    print("\n🧹 Ajustando nulos en LINEA y recalculando NUM_LINEAS...")
    
    # 1. Tipos correctos (Arrow / pandas nullable safe)
    df[col_linea] = df[col_linea].astype("string")
    df[col_num_lineas] = pd.to_numeric(df[col_num_lineas], errors="coerce")
    
    # 2. Normalizar texto
    df[col_linea] = df[col_linea].str.strip()
    
    # 3. Detectar valores que realmente deben ser nulos
    valores_nulos_texto = ["", " ", "NA", "N/A", "NULL", "None", "<NA>"]
    
    mask_linea_nula = (
        df[col_linea].isna() |
        df[col_linea].isin(valores_nulos_texto)
    )
    
    # 4. Convertir LINEA a nulo real (NULL)
    df.loc[mask_linea_nula, col_linea] = pd.NA
    
    # 5. Restar 1 a NUM_LINEAS donde LINEA es nula
    df.loc[mask_linea_nula, col_num_lineas] = (
        df.loc[mask_linea_nula, col_num_lineas] - 1
    )
    
    # 6. Regla de seguridad: si NUM_LINEAS <= 0 → NULL
    df.loc[df[col_num_lineas] <= 0, col_num_lineas] = pd.NA
    
    # ==========================
    # Validación
    # ==========================
    print("\n====== VALIDACIÓN FINAL ======")
    print("Registros donde LINEA es NULL:", mask_linea_nula.sum())
    print("Nulos finales en LINEA:", df[col_linea].isna().sum())
    print("Nulos finales en NUM_LINEAS:", df[col_num_lineas].isna().sum())
    
    print("\nEjemplo de registros ajustados:")
    print(df.loc[mask_linea_nula, [col_linea, col_num_lineas]].head())

else:
    print("⚠️ Columnas LINEA o NUM_LINEAS no encontradas")
```

```python
# ==========================
# CORRECCIÓN DEFINITIVA DE CONVERGENCIA (SIN <NA> NI NONE)
# ==========================

import pandas as pd
import numpy as np

col_conv2 = "CONVERGENCIA_2"
col_conv = "CONVERGENCIA"

if col_conv2 in df.columns and col_conv in df.columns:
    
    print("\n🧹 Iniciando corrección FINAL de CONVERGENCIA (clean business logic)...")
    
    # ==========================
    # 1️⃣ TIPOS SEGUROS
    # ==========================
    
    df[col_conv2] = df[col_conv2].astype("string").str.strip().str.upper()
    df[col_conv] = pd.to_numeric(df[col_conv], errors="coerce")
    
    # ==========================
    # 2️⃣ LIMPIEZA BASE (SIN_NIVEL)
    # ==========================
    
    # SIN_NIVEL → vacío
    df.loc[df[col_conv2] == "SIN_NIVEL", col_conv2] = ""
    
    # sufijos
    df[col_conv2] = df[col_conv2].str.replace(r"-SIN_NIVEL$", "", regex=True)
    
    # ==========================
    # 3️⃣ LIMPIEZA NONE (CRÍTICO)
    # ==========================
    
    # NONE puro
    df.loc[df[col_conv2] == "NONE", col_conv2] = ""
    
    # NONE en combinaciones
    df[col_conv2] = (
        df[col_conv2]
        .str.replace(r"\bNONE\b", "", regex=True)
        .str.replace(r"--+", "-", regex=True)
        .str.replace(r"^-|-$", "", regex=True)
        .str.strip()
    )


    # ==========================
# 3.5️⃣ LIMPIAR <NA> COMO TEXTO (CRÍTICO)
# ==========================

    df[col_conv2] = (
    df[col_conv2]
    .str.replace(r"\b<NA>\b", "", regex=True)   # elimina <NA>
    .str.replace(r"--+", "-", regex=True)       # limpia dobles guiones
    .str.replace(r"^-|-$", "", regex=True)      # limpia bordes
    .str.strip()
     )
    # ==========================
    # 4️⃣ ELIMINAR BASURA FINAL
    # ==========================
    
    # cualquier "<NA>" textual → vacío
    df[col_conv2] = df[col_conv2].replace("<NA>", "")
    
    # valores realmente nulos → vacío
    df[col_conv2] = df[col_conv2].fillna("")
    
    # ==========================
    # 5️⃣ FUNCIÓN DE NEGOCIO
    # ==========================
    
    servicios_validos = {"FIJO", "MOVIL", "DIGITAL"}
    
    def calcular_convergencia_real(x):
        if not x:  # vacío
            return 0
        
        servicios = {
            s.strip() for s in x.split("-")
            if s.strip() in servicios_validos
        }
        
        return len(servicios)
    
    # ==========================
    # 6️⃣ RECÁLCULO
    # ==========================
    
    df["CONVERGENCIA_CORREGIDA"] = (
        df[col_conv2]
        .apply(calcular_convergencia_real)
        .astype("Int8")
    )
    
    # ==========================
    # 7️⃣ AUDITORÍA
    # ==========================
    
    inconsistencias = (df[col_conv] != df["CONVERGENCIA_CORREGIDA"]).sum()
    print(f"\n🚨 Inconsistencias detectadas: {inconsistencias:,}")
    
    # reemplazo final
    df[col_conv] = df["CONVERGENCIA_CORREGIDA"]
    df.drop(columns=["CONVERGENCIA_CORREGIDA"], inplace=True)
    
    # ==========================
    # 8️⃣ VALIDACIONES FINALES
    # ==========================
    
    print("\n📊 Distribución FINAL:")
    print(df[col_conv].value_counts().sort_index())
    
    print("\n📊 Valores únicos finales CONVERGENCIA_2:")
    print(sorted(df[col_conv2].unique()))
    
    # 🔍 validaciones críticas
    
    print(f"\n🔎 Registros con NONE: {df[col_conv2].str.contains('NONE').sum()}")
    print(f"🔎 Registros con <NA>: {(df[col_conv2] == '<NA>').sum()}")
    print(f"🔎 Convergencia máxima: {df[col_conv].max()}")
    
else:
    print("⚠️ Columnas no encontradas")
```

```python
import pandas as pd

# ==========================
# 🔹 Reglas de negocio por SEGMENTO
# ==========================
qtile_rules = {
    "Small": 25,
    "SOHO": 15,
    "Wholesale International": 5,
    "Wholesale Others": 10,
    "Large": 10,
    "Gobierno": 10,
    "Empresas": 15,
    "Multinacionales": 10
}

# Límites globales
MAX_QTILES_RGU = 10
MAX_QTILES_LINEAS = 10
MIN_VALORES_POR_QTILE = 15

# ==========================
# 🔹 Función segura para qcut
# ==========================
def safe_qcut(series, n_qtiles):
    """qcut protegido: evita errores por pocos valores o duplicados"""
    if series.nunique() < 2 or n_qtiles < 2:
        return pd.Series(1, index=series.index)
    try:
        return pd.qcut(series, n_qtiles, labels=False, duplicates="drop") + 1
    except ValueError:
        n_final = min(series.nunique(), n_qtiles)
        if n_final < 2:
            return pd.Series(1, index=series.index)
        return pd.qcut(series, n_final, labels=False, duplicates="drop") + 1
```

```python
import pandas as pd

# ==========================
# 🔹 Reglas de negocio
# ==========================
qtile_rules = {
    "Small": 25,
    "SOHO": 15,
    "Wholesale International": 5,
    "Wholesale Others": 10,
    "Large": 10,
    "Gobierno": 10,
    "Empresas": 15,
    "Multinacionales": 10
}


MAX_QTILES = 10
MAX_QTILES_LINEAS = 10
MIN_VALORES_POR_QTILE = 15

# ==========================
# 🔹 Función segura
# ==========================
def safe_qcut(series, n_qtiles):
    if series.nunique() < 2 or n_qtiles < 2:
        return pd.Series(1, index=series.index)
    try:
        return pd.qcut(series, n_qtiles, labels=False, duplicates="drop") + 1
    except ValueError:
        n_qtiles_final = min(series.nunique(), n_qtiles)
        if n_qtiles_final < 2:
            return pd.Series(1, index=series.index)
        return pd.qcut(series, n_qtiles_final, labels=False, duplicates="drop") + 1

# ==========================
# 🔹 Qtiles por PERIODO y SEGMENTO
# ==========================
for (periodo, segmento), data in df_mes_limpio.groupby(["PERIODO", "SEGMENTO_2026"]):
    idx = data.index
    n_registros = len(data)

    # ---- FACTURACIÓN ----
    n_unicos = data["VALOR_FACTURADO_3M"].nunique()
    max_qtiles_regla = qtile_rules.get(segmento, 21)
    max_qtiles_volumen = n_registros // MIN_VALORES_POR_QTILE

    qtiles_fact = min(
        max_qtiles_regla,
        n_unicos,
        max_qtiles_volumen if max_qtiles_volumen >= 2 else 2
    )

    df_mes_limpio.loc[idx, "FACTURACION_PROMEDIO_QTILE"] = safe_qcut(
        data["VALOR_FACTURADO_3M"], qtiles_fact
    )

    # ---- RGU ----
    qtiles_rgu = min(MAX_QTILES, n_registros // MIN_VALORES_POR_QTILE)
    df_mes_limpio.loc[idx, "RGU_TOTAL_QTILE"] = safe_qcut(data["RGU_TOTAL"], max(qtiles_rgu, 2))

    # ---- LINEAS ----
    qtiles_lineas = min(MAX_QTILES_LINEAS, n_registros // MIN_VALORES_POR_QTILE)
    df_mes_limpio.loc[idx, "NUM_LINEA_QTILE"] = safe_qcut(data["NUM_LINEAS"], max(qtiles_lineas, 2))

    # ---- PRODUCTOS ----
    df_mes_limpio.loc[idx, "NUM_PRODUCTO_QTILE"] = safe_qcut(data["NUM_PRODUCTOS"], max(qtiles_lineas, 2))

    # ---- ANTIGÜEDAD ----
    df_mes_limpio.loc[idx, "MESES_ANTIGUEDAD_QTILE"] = safe_qcut(
        data["MESES_ANTIGUEDAD"], max(qtiles_rgu, 2)
    )

# ==========================
# 🔹 Resumen
# ==========================
resumen_qtiles = df_mes_limpio.groupby(["PERIODO", "SEGMENTO_2026"]).agg(
    FACTURACION_QTILE=("FACTURACION_PROMEDIO_QTILE", "max"),
    RGU_QTILE=("RGU_TOTAL_QTILE", "max"),
    ANTIGUEDAD_QTILE=("MESES_ANTIGUEDAD_QTILE", "max"),
    LINEAS_QTILE=("NUM_LINEA_QTILE", "max"),
    PRODUCTOS_QTILE=("NUM_PRODUCTO_QTILE", "max"),
    CLIENTES_UNICOS=("CLIENTE_ID", "nunique"),
    FACTURACION_TOTAL=("VALOR_FACTURADO_3M", "sum"),
    RGU_TOTAL=("RGU_TOTAL", "sum")
).reset_index()

# ==========================
# 🔹 Validación de totales
# ==========================
validacion = {
    "Base_Original": {
        "Facturacion_Total": df_mes_limpio["VALOR_FACTURADO_3M"].sum(),
        "Clientes_Unicos": df_mes_limpio["CLIENTE_ID"].nunique(),
        "RGU_Total": df_mes_limpio["RGU_TOTAL"].sum()
    },
    "Resumido_Qtiles": {
        "Facturacion_Total": resumen_qtiles["FACTURACION_TOTAL"].sum(),
        "Clientes_Unicos": resumen_qtiles["CLIENTES_UNICOS"].sum(),
        "RGU_Total": resumen_qtiles["RGU_TOTAL"].sum()
    }
}

# ==========================
# 🔹 Mostrar resultados
# ==========================
print("====== RESUMEN DE QTILES POR PERIODO Y SEGMENTO ======")
print(resumen_qtiles.to_string(index=False))

print("\n====== VALIDACIÓN DE TOTALES ======")
for k, v in validacion.items():
    print(f"\n-- {k} --")
    print(f"Facturación Total : {v['Facturacion_Total']:,.0f}")
    print(f"Clientes Únicos   : {v['Clientes_Unicos']:,}")
    print(f"RGU Total         : {v['RGU_Total']:,.0f}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ==========================
# 0️⃣ Asegurarse de que df_mes_limpio tenga las columnas QTILE generadas:
# FACTURACION_PROMEDIO_QTILE, RGU_TOTAL_QTILE, MESES_ANTIGUEDAD_QTILE, NUM_LINEA_QTILE
# ==========================
# Convertir a float para cálculos
df_mes_limpio['RANGO_FACTURACION'] = df_mes_limpio['FACTURACION_PROMEDIO_QTILE'].astype(float)
df_mes_limpio['RANGO_RGU'] = df_mes_limpio['RGU_TOTAL_QTILE'].astype(float)
df_mes_limpio['RANGO_ANTIGUEDAD'] = df_mes_limpio['MESES_ANTIGUEDAD_QTILE'].astype(float)
df_mes_limpio['RANGO_LINEA'] = df_mes_limpio['NUM_LINEA_QTILE'].astype(float)

# ==========================
# 1️⃣ Calcular CLV ponderado
# ==========================
PESO_FACT = 0.50
PESO_RGU = 0.15
PESO_ANTIG = 0.00
PESO_LINEA = 0.35

df_mes_limpio['CLV'] = (
    df_mes_limpio['RANGO_FACTURACION']*PESO_FACT +
    df_mes_limpio['RANGO_RGU']*PESO_RGU +
    df_mes_limpio['RANGO_ANTIGUEDAD']*PESO_ANTIG +
    df_mes_limpio['RANGO_LINEA']*PESO_LINEA
)

# ==========================
# 2️⃣ Clasificación por percentil dentro de cada segmento
# ==========================
def clasifica_clv_percentil(df_segmento, pct_blindaje=0.85, pct_rent=0.60):
    clv_85 = df_segmento['CLV'].quantile(pct_blindaje)
    clv_60 = df_segmento['CLV'].quantile(pct_rent)
    
    def clasifica(row):
        if row['CLV'] >= clv_85:
            return 'Blindaje'
        elif row['CLV'] >= clv_60:
            return 'Rentabilización'
        else:
            return 'Penetración'
    
    df_segmento['CLV_GRUPO'] = df_segmento.apply(clasifica, axis=1)
    return df_segmento

df_mes_limpio = df_mes_limpio.groupby('SEGMENTO_2026', group_keys=False).apply(clasifica_clv_percentil)

# ==========================
# 3️⃣ Resumen de qtiles representativos por grupo CLV
# ==========================
qtile_cols = ['FACTURACION_PROMEDIO_QTILE','RGU_TOTAL_QTILE','MESES_ANTIGUEDAD_QTILE','NUM_LINEA_QTILE']
resumen_qtile_grupo = df_mes_limpio.groupby('CLV_GRUPO')[qtile_cols].agg(lambda x: x.mode()[0]).reset_index()

print("====== Qtiles representativos por grupo CLV ======")
print(resumen_qtile_grupo.to_string(index=False))

# ==========================
# 4️⃣ Graficar distribución de qtiles por grupo CLV
# ==========================
for var in qtile_cols:
    plt.figure(figsize=(8,5))
    sns.countplot(x=var, hue='CLV_GRUPO', data=df_mes_limpio, palette='Set2', order=sorted(df_mes_limpio[var].unique()))
    plt.title(f'Distribución de {var} por grupo CLV')
    plt.ylabel('Cantidad de clientes')
    plt.xlabel('Qtile / Nivel')
    plt.legend(title='CLV Grupo')
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd

# ============================================================
# 0️⃣ COPIA BASE (CONTROL DE INTEGRIDAD)
# ============================================================
df_mes_limpio_base = df_mes_limpio.copy()

# ============================================================
# 1️⃣ LIMPIEZA DE EJECUTIVOS (df9)
# ============================================================

df9['Número de Identificación'] = (
    df9['Número de Identificación']
    .astype(str)
    .str.replace(r'\.0$', '', regex=True)
    .str.strip()
)

df9['Dígito de verificación'] = (
    df9['Dígito de verificación']
    .astype(str)
    .str.replace(r'\.0$', '', regex=True)
    .str.strip()
)

df9['Dígito de verificación'] = df9['Dígito de verificación'].replace(
    ['nan', 'NaN', 'None', ''], ''
)

# Crear NIT (CLAVE_ID)
df9['NIT'] = df9.apply(
    lambda x: x['Número de Identificación']
    if x['Dígito de verificación'] == ''
    else f"{x['Número de Identificación']}-{x['Dígito de verificación']}",
    axis=1
)

df9['CLAVE_ID'] = df9['NIT'].astype(str).str.strip()

# ============================================================
# 2️⃣ DATAFRAME DE EJECUTIVOS (SIN DUPLICADOS)
# ============================================================

df_exec = (
    df9[[
        'CLAVE_ID',
        'Propietario del cliente: Nombre completo',
        'Propietario del cliente: Gestor: Nombre completo'
    ]]
    .rename(columns={
        'Propietario del cliente: Nombre completo': 'EJECUTIVO_NEW',
        'Propietario del cliente: Gestor: Nombre completo': 'GESTOR_NEW'
    })
    .dropna(subset=['CLAVE_ID'])
    .drop_duplicates(subset='CLAVE_ID')
)

# ============================================================
# 3️⃣ PREPARAR df_mes_limpio (TIPOS CONSISTENTES)
# ============================================================

df_mes_limpio['CLIENTE_ID'] = (
    df_mes_limpio['CLIENTE_ID']
    .astype(str)
    .str.strip()
)

# ============================================================
# 4️⃣ MERGE PRINCIPAL POR CLIENTE_ID (CONTROLADO)
# ============================================================

filas_antes = len(df_mes_limpio)

df_final_merge = df_mes_limpio.merge(
    df_exec,
    how='left',
    left_on='CLIENTE_ID',
    right_on='CLAVE_ID',
    validate='m:1'  # evita duplicación silenciosa
)

# Flag de match por ID
df_final_merge['_MATCH_ID'] = df_final_merge['EJECUTIVO_NEW'].notna()

# ============================================================
# 5️⃣ CREAR / ACTUALIZAR EJECUTIVO Y GESTOR (PRIORIDAD: ID)
# ============================================================

if 'EJECUTIVO' not in df_final_merge.columns:
    df_final_merge['EJECUTIVO'] = pd.NA

if 'GESTOR' not in df_final_merge.columns:
    df_final_merge['GESTOR'] = pd.NA

df_final_merge['EJECUTIVO'] = df_final_merge['EJECUTIVO_NEW'].combine_first(
    df_final_merge['EJECUTIVO']
)

df_final_merge['GESTOR'] = df_final_merge['GESTOR_NEW'].combine_first(
    df_final_merge['GESTOR']
)

# ============================================================
# 6️⃣ MERGE POR NOMBRE (SOLO SIN MATCH POR ID)
# ============================================================

df_final_merge['_NORM_NOMBRE'] = (
    df_final_merge['CLIENTE']
    .astype(str)
    .str.strip()
    .str.lower()
)

df9['_NORM_NOMBRE'] = (
    df9['Nombre del cliente']
    .astype(str)
    .str.strip()
    .str.lower()
)

sin_match_id = df_final_merge[
    df_final_merge['EJECUTIVO'].isna() &
    df_final_merge['GESTOR'].isna()
].copy()

df_exec_nombre = (
    df9[[
        '_NORM_NOMBRE',
        'Propietario del cliente: Nombre completo',
        'Propietario del cliente: Gestor: Nombre completo'
    ]]
    .rename(columns={
        'Propietario del cliente: Nombre completo': 'EJECUTIVO_NOMBRE',
        'Propietario del cliente: Gestor: Nombre completo': 'GESTOR_NOMBRE'
    })
    .drop_duplicates(subset='_NORM_NOMBRE')
)

sin_match_id = sin_match_id.merge(
    df_exec_nombre,
    how='left',
    on='_NORM_NOMBRE'
)

# Actualizar SOLO donde encontró por nombre
mask_nombre = sin_match_id['EJECUTIVO_NOMBRE'].notna()

df_final_merge.loc[sin_match_id.index[mask_nombre], 'EJECUTIVO'] = (
    sin_match_id.loc[mask_nombre, 'EJECUTIVO_NOMBRE']
)

df_final_merge.loc[sin_match_id.index[mask_nombre], 'GESTOR'] = (
    sin_match_id.loc[mask_nombre, 'GESTOR_NOMBRE']
)

# Flag match por nombre
df_final_merge['_MATCH_NOMBRE'] = False
df_final_merge.loc[sin_match_id.index[mask_nombre], '_MATCH_NOMBRE'] = True

# ============================================================
# 7️⃣ LIMPIEZA COLUMNAS AUXILIARES
# ============================================================

df_final_merge.drop(
    columns=[
        'EJECUTIVO_NEW', 'GESTOR_NEW', 'CLAVE_ID',
        '_NORM_NOMBRE'
    ],
    inplace=True,
    errors='ignore'
)

df9.drop(columns=['_NORM_NOMBRE'], inplace=True, errors='ignore')

# ============================================================
# 8️⃣ VALIDACIONES DE INTEGRIDAD (CRÍTICAS)
# ============================================================

print("\n========== VALIDACIÓN DE FILAS ==========")
print("Filas df_mes_limpio :", filas_antes)
print("Filas df_final_merge:", len(df_final_merge))

assert filas_antes == len(df_final_merge), "❌ ERROR: Se perdieron o duplicaron filas"

print("\n========== VALIDACIÓN DE TOTALES ==========")
print("FACT ORIGINAL :", df_mes_limpio_base['VALOR_FACTURACION'].sum())
print("FACT FINAL    :", df_final_merge['VALOR_FACTURACION'].sum())

print("RGU ORIGINAL  :", df_mes_limpio_base['RGU_TOTAL'].sum())
print("RGU FINAL     :", df_final_merge['RGU_TOTAL'].sum())

# ============================================================
# 9️⃣ RESÚMENES COMPLETOS DE CALIDAD DE MATCH
# ============================================================

total_registros = len(df_final_merge)
match_id = df_final_merge['_MATCH_ID'].sum()
match_nombre = df_final_merge['_MATCH_NOMBRE'].sum()
sin_match_total = (
    df_final_merge['EJECUTIVO'].isna() &
    df_final_merge['GESTOR'].isna()
).sum()

print("\n========== RESUMEN DE MATCH ==========")
print(f"Total registros        : {total_registros:,}")
print(f"Match por CLIENTE_ID   : {match_id:,}")
print(f"Match por NOMBRE       : {match_nombre:,}")
print(f"Sin match (final)      : {sin_match_total:,}")

print("\n% Cobertura ejecutivos:",
      f"{((total_registros - sin_match_total) / total_registros * 100):.2f}%")

# ============================================================
# 🔟 RESUMEN FINANCIERO FINAL POR PERIODO
# ============================================================

resumen_periodo = (
    df_final_merge
    .groupby('PERIODO', as_index=False)[['VALOR_FACTURACION', 'RGU_TOTAL']]
    .sum()
    .sort_values('PERIODO')
)

resumen_periodo['VALOR_FACTURACION'] = resumen_periodo['VALOR_FACTURACION'].apply(
    lambda x: f"{x:,.0f}".replace(",", ".")
)

resumen_periodo['RGU_TOTAL'] = resumen_periodo['RGU_TOTAL'].apply(
    lambda x: f"{x:,.0f}".replace(",", ".")
)

print("\n📊 SUMA FINAL POR PERIODO (100% CONSERVADA):\n")
print(resumen_periodo.to_string(index=False))

# ============================================================
# 1️⃣1️⃣ PRINT FINAL DE COLUMNAS (LO QUE PEDISTE)
# ============================================================

print("\n====== COLUMNAS FINALES DEL DATAFRAME ======")
print(df_final_merge.columns.tolist())
```

```python
import pandas as pd
import numpy as np

# ==============================================================
# 🧱 0. BASE SEGURA
# ==============================================================

# Copia para no afectar el modelo original
df_powerbi = df_mes_limpio.copy()

print(f"📦 Registros iniciales: {len(df_powerbi):,}")

# ==============================================================
# 🧹 1. LIMPIEZA CRÍTICA DE CLIENTE_ID (SIN ELIMINAR CLIENTES)
# ==============================================================

df_powerbi["CLIENTE_ID"] = (
    df_powerbi["CLIENTE_ID"]
    .astype(str)
    .str.strip()
)

# Convertir textos inválidos a NaN (pero NO eliminar filas)
ids_invalidos = ["", "nan", "none", "null", "tbd", "NaN", "None", "NULL", "TBD"]
df_powerbi.loc[
    df_powerbi["CLIENTE_ID"].str.lower().isin(ids_invalidos),
    "CLIENTE_ID"
] = pd.NA

print(f"🧾 CLIENTE_ID nulos/vacíos detectados: {df_powerbi['CLIENTE_ID'].isna().sum():,}")

# ==============================================================
# 📅 2. TIPOS DE DATOS (SIN FILTRAR REGISTROS)
# ==============================================================

df_powerbi["PERIODO"] = pd.to_datetime(df_powerbi["PERIODO"], errors="coerce")

cols_numericas = [
    "VALOR_FACTURACION",
    "VALOR_FACTURADO_3M",
    "RGU_TOTAL",
    "NUM_LINEAS",
    "NUM_PRODUCTOS",
    "MESES_ANTIGUEDAD"
]

for col in cols_numericas:
    if col in df_powerbi.columns:
        df_powerbi[col] = pd.to_numeric(df_powerbi[col], errors="coerce")

print("🔢 Conversión numérica completada (sin eliminación de filas).")

# ==============================================================
# 🚨 3. VALIDACIONES INTELIGENTES (NO ELIMINAR CLIENTES)
# ==============================================================

# 3.1 Valores negativos → NaN (NO eliminar)
for col in ["VALOR_FACTURACION", "RGU_TOTAL", "VALOR_FACTURADO_3M"]:
    if col in df_powerbi.columns:
        negativos = (df_powerbi[col] < 0).sum()
        if negativos > 0:
            print(f"⚠️ {negativos:,} valores negativos en {col} convertidos a NaN (no eliminados).")
            df_powerbi.loc[df_powerbi[col] < 0, col] = np.nan


# ==============================================================
# 🧠 4. IMPUTACIÓN CONSERVADORA (CLAVE PARA NO PERDER CLIENTES)
# ==============================================================

# Nunca eliminar clientes por NaN operativos
fill_zero_cols = [
    "VALOR_FACTURACION",
    "VALOR_FACTURADO_3M",
    "RGU_TOTAL",
    "NUM_LINEAS",
    "NUM_PRODUCTOS"
]

for col in fill_zero_cols:
    if col in df_powerbi.columns:
        n_nan = df_powerbi[col].isna().sum()
        if n_nan > 0:
            print(f"🔧 Imputando {n_nan:,} NaN en {col} con 0 (Power BI safe).")
            df_powerbi[col] = df_powerbi[col].fillna(0)

# ==============================================================
# 📋 5. VALIDACIÓN DE CAMPOS CATEGÓRICOS
# ==============================================================

if "EJECUTIVO" in df_powerbi.columns:
    faltantes_ejecutivo = df_powerbi["EJECUTIVO"].isna().sum()
    print(f"📋 Valores faltantes en EJECUTIVO: {faltantes_ejecutivo:,}")

# ==============================================================
# 🔍 6. AUDITORÍA: ¿SE ESTÁN PERDIENDO CLIENTES?
# ==============================================================

clientes_original = (
    df_mes_limpio.groupby("PERIODO")["CLIENTE_ID"]
    .nunique(dropna=True)
    .reset_index(name="CLIENTES_ORIGINALES")
)

clientes_powerbi = (
    df_powerbi.groupby("PERIODO")["CLIENTE_ID"]
    .nunique(dropna=True)
    .reset_index(name="CLIENTES_POWERBI")
)

control_clientes = clientes_original.merge(
    clientes_powerbi,
    on="PERIODO",
    how="left"
)

control_clientes["DIFERENCIA"] = (
    control_clientes["CLIENTES_ORIGINALES"] -
    control_clientes["CLIENTES_POWERBI"]
)

print("\n🧾 CONTROL DE PÉRDIDA DE CLIENTES:")
print(control_clientes.to_string(index=False))

# ==============================================================
# 📊 7. RESUMEN DE CONSISTENCIA PARA POWER BI
# ==============================================================

resumen_periodo = (
    df_powerbi.groupby("PERIODO")
    .agg(
        CLIENTES_UNICOS=("CLIENTE_ID", lambda x: x.nunique(dropna=True)),
        REGISTROS=("CLIENTE_ID", "size"),
        IDS_NULOS=("CLIENTE_ID", lambda x: x.isna().sum()),
        TOTAL_FACTURACION=("VALOR_FACTURACION", "sum"),
        TOTAL_RGU=("RGU_TOTAL", "sum"),
        PROM_LINEAS=("NUM_LINEAS", "mean"),
        PROM_PRODUCTOS=("NUM_PRODUCTOS", "mean"),
    )
    .reset_index()
)

# Formato visual
resumen_periodo["TOTAL_FACTURACION"] = resumen_periodo["TOTAL_FACTURACION"].apply(
    lambda x: f"${x:,.0f}".replace(",", ".") if pd.notna(x) else "-"
)

resumen_periodo["TOTAL_RGU"] = resumen_periodo["TOTAL_RGU"].apply(
    lambda x: f"{x:,.0f}".replace(",", ".") if pd.notna(x) else "-"
)

print("\n📊 Resumen de consistencia por PERIODO (Power BI Ready):")
print(resumen_periodo.to_string(index=False))

# ==============================================================
# 💾 8. EXPORTACIÓN FINAL OPTIMIZADA PARA POWER BI
# ==============================================================

columnas_export = [c for c in [
    'PERIODO', 'CLIENTE_ID', 'CLIENTE',
    'VALOR_FACTURACION', 'VALOR_FACTURADO_3M',
    'RGU_TOTAL', 'NUM_LINEAS', 'NUM_PRODUCTOS',
    'MESES_ANTIGUEDAD', 'EJECUTIVO', 'SEGMENTO',
    'SEGMENTO_2026', 'REGIONAL', 'DEPARTAMENTO',
    'MUNICIPIO', 'LINEA', 'PRODUCTO_H',
    'CONVERGENCIA', 'CONVERGENCIA_2',
    'FIRST_LEVEL', 'SECOND_LEVEL',
    'CLV', 'CLV_GRUPO', 'GESTOR'
] if c in df_powerbi.columns]

df_powerbi[columnas_export].to_csv(
   "BASE_B2B_CLV_32.csv",
    index=False,
    encoding="utf-8-sig"
)

# ==============================================================
# ✅ 9. RESUMEN FINAL
# ==============================================================

print("\n✅ DATASET FINAL LISTO PARA POWER BI")
print(f"📦 Registros finales: {df_powerbi.shape[0]:,}")
print(f"📅 Períodos únicos: {df_powerbi['PERIODO'].nunique()}")
print(f"👥 Clientes únicos totales: {df_powerbi['CLIENTE_ID'].nunique():,}")
print(f"📁 Archivo exportado: BASE_B2B_CLV_POWERBI_FINAL.csv")
```
